## Descriptive Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# reading data
data=pd.read_csv("covid_19_data.csv")

# convert data into dataframe
df=pd.DataFrame(data)

# exploring first 10 entries
df.head(10)
```

```python
# to change the name of attributes
df=df.rename(columns={"Province/State":"Province","Country/Region":"Country","ObservationDate":"Date"})

df.head()
```

```python
# to get statistics of the dataset
print(df.describe())

print("\n")

# to get information of attributes and entries
df.info()
```

```python
# Exploring last 20 entries 
df.tail(20)
```

```python
# to check total missing values column-wise
df.isnull().sum()
```

```python
# to arrange the dataframe country-wise and to get sum of all confirmed, deaths and recovered cases
rf = df.groupby(['Country','Date'])[['Country','Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
rf.head()
```

```python
# Arranging data in descending order as per confirmed cases
sorted_By_Confirmed=rf.sort_values('Confirmed',ascending=False)
sorted_By_Confirmed.head(10)
```

```python
# Droping duplicates entries of countries
sorted_By_Confirmed=sorted_By_Confirmed.drop_duplicates('Country')
sorted_By_Confirmed.tail(40)
```

```python
world_Confirmed_Total=sorted_By_Confirmed['Confirmed'].sum()

world_Deaths_Total=sorted_By_Confirmed['Deaths'].sum()

world_Recovered_Total=sorted_By_Confirmed['Recovered'].sum()

active = world_Confirmed_Total - world_Deaths_Total - world_Recovered_Total 

# Creating dictionary
veri={'Total Confirmed cases  in the world':world_Confirmed_Total,
      'Total Deaths cases in the world':world_Deaths_Total,
      'Total Recovered cases in the world':world_Recovered_Total,
      'Total Active Cases':active}

print("The data is till 21st March 2020:")
veri=pd.DataFrame.from_dict(veri, orient='index' ,columns=['Total'])
veri.style.background_gradient(cmap='Greens')
```

```python
Recovered_rate=(sorted_By_Confirmed['Recovered']*100)/sorted_By_Confirmed['Confirmed']
Deaths_rate=(sorted_By_Confirmed['Deaths']*100)/sorted_By_Confirmed['Confirmed']
cases_rate=(sorted_By_Confirmed.Confirmed*100)/world_Confirmed_Total

sorted_By_Confirmed['Active']=sorted_By_Confirmed['Confirmed']-sorted_By_Confirmed['Deaths']-sorted_By_Confirmed['Recovered']
sorted_By_Confirmed['Recovered Cases Rate %']=pd.DataFrame(Recovered_rate)
sorted_By_Confirmed['Deaths Cases Rate %']=pd.DataFrame(Deaths_rate)
sorted_By_Confirmed['Total Cases Rate %']=pd.DataFrame(cases_rate)


print("Sorted By Confirmed Cases:")
sorted_By_Confirmed.style.background_gradient(cmap='Blues')
```

## Data Visualization

### 1. Comparison of Confirmed, Recovered and Death cases of different countries

```python
sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(15,15 ))

sns.barplot(x="Confirmed", y="Country", data=sorted_By_Confirmed.head(20),
            label="Confirmed", color="b")

sns.set_color_codes("muted")
sns.barplot(x="Recovered", y="Country", data=sorted_By_Confirmed.head(20),
            label="Recovered", color="g")

sns.set_color_codes("muted")
sns.barplot(x="Deaths", y="Country", data=sorted_By_Confirmed.head(20),
            label="Deaths", color="r")

# Add a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
```

### 2. Top 10 countries in the world facing Covid-19

### 3. Number of confirmed, recovered and death cases per day

```python
cases_per_Day = df.groupby(["Date"])['Confirmed','Deaths', 'Recovered'].sum().reset_index()

sorted_By_Confirmed1 = cases_per_Day.sort_values('Date',ascending=False)

x=cases_per_Day.index

y=cases_per_Day.Confirmed
y1=cases_per_Day.Deaths
y2=cases_per_Day.Recovered

sns.set(style="whitegrid")

# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(12,10 ))

plt.scatter(x,y,color='blue' ,label='Confirmed Cases')
plt.scatter(x,y1,color='red' ,label="Deaths Cases")
plt.scatter(x,y2,color='green',label="Recovered Cases")
plt.title("Increasing infections cases in the world per day")
ax.legend(ncol=2, loc='upper left', frameon=True)
plt.show()
```

## Predictive Analysis

### 1. Feature Selection and Training of Polynomial Regression Model

```python
# feature selection 
x_data=pd.DataFrame(cases_per_Day.index)   # prediction variable
y_data=pd.DataFrame(cases_per_Day.Confirmed)  # target variable

# Now, the features is to be divided into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data,test_size=0.1,random_state=0)
```

#### Let us first analyse the data with Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lin_reg1=LinearRegression()

# Training of linear regression model
lin_reg1.fit(x_train,y_train)

# Prediction from linear regression model
lin_reg1.predict(x_test)
```

```python
# Visualising the Linear Regression results 
plt.scatter(x_test, y_test, color = 'blue') 
  
plt.plot(x_test, lin_reg1.predict(x_test), color = 'red') 
plt.title('Linear Regression') 
plt.xlabel('No. of days') 
plt.ylabel('No. of Confirmed Cases') 
  
plt.show() 
```

#### Since linear regression is not suitable as it is leaving outlier which has a significant value and can not be neglected therefore I will go for Polynomial Regression which is preferred for this kind of dataset. 

```python
# importing Polynomial features from sci-kit learn library
from sklearn.preprocessing import PolynomialFeatures

# Adding polynomial features in training data
poly_reg=PolynomialFeatures(degree=5)
x_poly=poly_reg.fit_transform(x_train)

lin_reg2=LinearRegression()
lin_reg2.fit(x_poly,y_train)
```

### 2. Making Prediction With Polynomial Regression Model

#### Since the model has seccessfuly trained. Now, this is the time to predict the number of confirmed cases per day in the world.

```python
# Making prediction using test data
y_pred=lin_reg2.predict(poly_reg.fit_transform(x_test))

# initialization of result by storing predicted value as DataFrame
result=pd.DataFrame(y_pred)
result['Real Value']=y_test.iloc[:,:].values

# Converting predicted values into dataframe
result['Predicted Value']=pd.DataFrame(y_pred)

# Side-by-side comparison of Real Target Values and Predicted Target Values
result=result[['Real Value','Predicted Value']]
result
```

#### Visualization of predicted value by Polynomial Regression Model with test data

```python
plt.scatter(x_test,y_test,color='blue') 
plt.plot(x_test,y_pred,color='red')
plt.title("Polynomial Regression Model") 
plt.xlabel('No. of days') 
plt.ylabel('No. of Confirmed Cases') 
plt.show()
```

#### Let us check the accuracy of the model

```python
# to check the accuracy import r2_score from metrics
from sklearn.metrics import r2_score

print('Polynomial Regession  R2 Score:',r2_score(y_test, y_pred))
```

##### Hence, the model is working with good accuracy.
